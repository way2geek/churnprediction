```python
import numpy as np
import pandas as pd
import copy
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
dataset = pd.read_csv('./BankChurners.csv')
df = copy.deepcopy(dataset)
df
```

## Data Recoding 

```python
Attrition_Flag = ['Attrited Customer', 'Existing Customer']
Income_Category = ['Less than $40K','$40K - $60K','$60K - $80K','$80K - $120K', '$120K +','Unknown']
Education_Level = ['Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate', 'Unknown']


attrition = LabelEncoder()
edu = OrdinalEncoder(categories=[Education_Level],
                     handle_unknown='use_encoded_value', 
                     unknown_value=np.nan)
income = OrdinalEncoder(categories=[Income_Category],
                     handle_unknown='use_encoded_value', 
                     unknown_value=np.nan)


df['Attrition_Flag'] = attrition.fit_transform(df['Attrition_Flag'])
X = edu.fit_transform(df[['Education_Level']])
df['Education_Level'] = [i for arr in X for i in arr]
Y = income.fit_transform(df[['Income_Category']])
df['Income_Category'] = [i for arr in X for i in arr]
```

## Replacing the 'Unknown' value with the meadian of the 'edu' and 'income' column

```python
df['Income_Category'] = df['Income_Category'].replace(5, np.median(df['Income_Category']))
df['Education_Level'] = df['Education_Level'].replace(6, np.median(df['Income_Category']))
```

## Encoding categorical data

```python
pd.get_dummies(df.Gender)
pd.get_dummies(df.Marital_Status)
pd.get_dummies(df.Card_Category)

df = df.join(pd.get_dummies(df.Gender))
df = df.join(pd.get_dummies(df.Marital_Status))
df = df.join(pd.get_dummies(df.Card_Category))

df.drop(['CLIENTNUM','Gender','Marital_Status','Card_Category'], axis=1, inplace=True)
df
```

## Statistical Features

```python
df.describe().apply(lambda x :round(x,3))
```

## Standardization

```python
df = (df - df.min()) / (df.max() - df.min())
df.apply(lambda x :round(x,4))
```
