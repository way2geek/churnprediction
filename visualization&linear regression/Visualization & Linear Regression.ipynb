# Linear Regression of Numerical Dataset

```python
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.model_selection import KFold
from sklearn.decomposition import PCA


df_processed = pd.read_csv('BankChurners_processed.csv')
df_numerical_processed = df_processed.loc[:,df_processed.dtypes != np.object]


col = df_numerical.columns.values.tolist()
col.remove('Credit_Limit')

X = df_numerical[col]
y= df_numerical['Credit_Limit']
score=0
n=5
kf = KFold(n_splits=n, shuffle=False)
for train_index, test_index in kf.split(X):
    X_train, X_test = X.values[train_index], X.values[test_index]
    y_train, y_test = y.values[train_index], y.values[test_index]
    lr = LinearRegression().fit(X_train, y_train)
    print('The coefficients:')
    print(lr.coef_)
    print('The intercept:')
    print(lr.intercept_)
    score+=lr.score(X_train,y_train)
print('The R-squared is ',score/n,', it means ','{0:.0f}%'.format(score/n * 100),'Credit Limit can be explained by this model')
   
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]

print(vif_data)
```

```python
need_to_remove = ['Customer_Age','Months_on_book','Total_Amt_Chng_Q4_Q1','Total_Trans_Ct'] # Remove parameters which VIF > 10
for i in need_to_remove:
    col.remove(i)
X = df_numerical[col]
y= df_numerical['Credit_Limit']
score=0
n=2
kf = KFold(n_splits=n, shuffle=False)
for train_index, test_index in kf.split(X):
    X_train, X_test = X.values[train_index], X.values[test_index]
    y_train, y_test = y.values[train_index], y.values[test_index]
    lr = LinearRegression().fit(X_train, y_train)
    print(lr.coef_)
    print(lr.intercept_)
    score+=lr.score(X_train,y_train)
print('The R-squared is ',score/n,', it means ','{0:.0f}%'.format(score/n * 100),'Credit Limit can be explained by this model')
   
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]

print(vif_data)
```
