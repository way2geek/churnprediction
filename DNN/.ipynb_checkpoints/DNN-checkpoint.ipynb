## Model Building

```python
class dnn(nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(input_dim,30),    
            nn.ReLU(),  
            nn.Linear(30,20),
            nn.ReLU(), 
            nn.Linear(20,20),
            nn.ReLU(), 
            nn.Linear(20,15),
            nn.ReLU(), 
            nn.Linear(15,output_dim),
            nn.Sigmoid()
            )
    def forward(self, x):
        x = x.view(x.shape[0], -1)
        return self.layers(x) 
```

```python
dnn = dnn(26,1)
```

```python
loss = torch.nn.BCELoss()
optimizer = torch.optim.SGD(dnn.parameters(),
                            lr=0.01,   #0.5,0.1,0.05,0.01
                            weight_decay=0.002,  
                            momentum=0.9)
#optimizer = torch.optim.Adam(dnn.parameters(), lr=0.01)
#optimizer = torch.optim.Adagrad(dnn.parameters(), lr=0.01)
```

```python
class ChurnDataset(torch.utils.data.Dataset):
    def __init__(self,churnpd):
        churn=churnpd.copy()
 
        x=churn[Churn.columns[1:]].values
        y=churn.Attrition_Flag.values
 
        self.x_train=torch.tensor(x,dtype=torch.float32)
        self.y_train=torch.tensor(y,dtype=torch.float32)

    def __len__(self):
            
        return len(self.y_train)

    def __getitem__(self,idx):
        return self.x_train[idx],self.y_train[idx]
```

## Model Evaluation

```python
from torchmetrics.classification import BinaryF1Score
from torchmetrics.classification import BinaryRecall
from torchmetrics.classification import BinaryAUROC
from torchmetrics.classification import BinaryAccuracy
from torchmetrics.classification import BinaryPrecision
from torchmetrics.classification import BinaryROC
from torchmetrics.classification import BinaryConfusionMatrix
F1 = BinaryF1Score()
Rec = BinaryRecall()
AUROC = BinaryAUROC()
Acc  = BinaryAccuracy()
Prec = BinaryPrecision()
ROC = BinaryROC()
Con_Mat = BinaryConfusionMatrix()
ev = pd.DataFrame(columns = ['epoch','Accuracy','Recall','Precision','F1_score','AUROC','time'])
```

```python
def plotroc(fpr,tpr,auroc):    
    plt.figure()
    lw = 2
    plt.figure(figsize=(10,10))
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC CURVE')
    plt.plot(fpr.tolist(), tpr.tolist(), color='orange',lw=lw, label='ROC curve (area = %0.3f)' % auroc)
    plt.show()
    return plt
def plotcm(mat):
    plt.matshow(mat.numpy(), cmap=plt.cm.Greens)  
    plt.colorbar()  
    plt.ylabel('True label')  
    plt.xlabel('Predicted label')
    plt.show()
def evaluation(preds,targets,preds_prob,tick):
    f1 = F1(preds,targets).tolist()
    rec = Rec(preds,targets).tolist()
    auroc = AUROC(preds,targets).tolist()
    acc  = Acc(preds,targets).tolist()
    prec = Prec(preds,targets).tolist()
    fpr, tpr, thresholds = ROC(preds_prob,targets)
    con_mat = Con_Mat(preds,targets)
    tok  = time.time()
    print('*'*80)
    ev.loc[epoch] = [epoch,round(acc,4),round(rec,4),round(prec,4),round(f1,4),round(auroc,4),round((tok-tick),4)]
    print(ev)
    plotroc(fpr,tpr,auroc)
    plotcm(con_mat)    
```

## Training&Validation&Tesing

```python
min_valid_loss = np.inf
begin= time.time()
for epoch in range(30):
    preds = torch.tensor([])
    targets = torch.tensor([])
    preds_prob =  torch.tensor([])
    tick = time.time()
    kf5 = KFold(n_splits=10, shuffle=True)
    for train, test in kf5.split(Churn):
        dnn.train() 
        trainset = Churn.iloc[train].copy()
        testset  = Churn.iloc[test].copy()
        trDataset = ChurnDataset(trainset)
        teDataset = ChurnDataset(testset)
        train, valid = torch.utils.data.random_split(trDataset,[trainset.shape[0]-int(trainset.shape[0]/5),int(trainset.shape[0]/5)])
        validloader = torch.utils.data.DataLoader(valid, batch_size=128, shuffle=True)
        trainloader = torch.utils.data.DataLoader(trDataset, batch_size=128, shuffle=True)
        testloader =  torch.utils.data.DataLoader(teDataset, batch_size=128, shuffle=True)
        for x,y in trainloader:
            optimizer.zero_grad() 
            output = dnn(x)  
            new_loss = loss(output.view(-1), y)  
            new_loss.backward() 
            optimizer.step()
        dnn.eval()
        valid_loss = 0.0
        for data, labels in validloader:            
            target = dnn(data)
            lossvalue = loss(target.view(-1),labels)
            valid_loss += lossvalue.item()        
        if min_valid_loss > valid_loss:            
            min_valid_loss = valid_loss
            torch.save(dnn.state_dict(), 'best_model')
        dnn.eval() 
        torch.load('best_model')
        with torch.no_grad():
            for x,y in testloader:
                output = dnn(x) 
                preds_prob =torch.cat((preds_prob,output.view(-1)))
                output[output>0.5] = 1
                output[output<=0.5] = 0
                res_lbl = output.view(-1)
                preds = torch.cat((preds,res_lbl))
                targets = torch.cat((targets,y))
    evaluation(preds,targets,preds_prob,tick)
end= time.time()
print(f'total cost {(end-begin)} s')
```
