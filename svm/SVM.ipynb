```python
import numpy as np
import pandas as pd
import copy
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.svm import SVC
```

```python
dataset = pd.read_csv('./BankChurners.csv')
df = copy.deepcopy(dataset)
df
```

```python
col_1 = df['Attrition_Flag']
col_12 = df.iloc[:,2:29]
x= pd.DataFrame(col_12)  
y= pd.DataFrame(col_1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
kf = KFold(n_splits=5, shuffle=True, random_state=15)
accuracy_score_rbf=[]
precision_rbf = []
recall_score_rbf=[]
f1_score_rbf=[]
for train_index, test_index in kf.split(df):
    X_train, X_test = x.values[train_index], x.values[test_index]
    y_train, y_test = y.values[train_index], y.values[test_index]
    clf = SVC(kernel = 'rbf', random_state = 0)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy_score_rbf.append(accuracy_score(y_test, y_pred))
    accuracy_score_rbf = [round(i,4) for i in accuracy_score_rbf]
    precision_rbf.append(precision_score(y_test, y_pred))
    precision_rbf = [round(i,4) for i in precision_rbf]
    recall_score_rbf.append(recall_score(y_test, y_pred))
    recall_score_rbf = [round(i,4) for i in recall_score_rbf]
    f1_score_rbf.append(f1_score(y_test, y_pred))
    f1_score_rbf = [round(i,4) for i in f1_score_rbf]
print('The accury score of SVM with Guassian Kernel based on 5-fold cross validation:\n',accuracy_score_rbf)
print('The precision of SVM with Guassian Kernel based on 5-fold cross validation:\n',precision_rbf)
print('The recall score of SVM with Guassian Kernel based on 5-fold cross validation:\n', recall_score_rbf)
print('The f1 score of SVM with Guassian Kernel based on 5-fold cross validation:\n', f1_score_rbf)
```
