```python
import numpy as np
import pandas as pd
import copy
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.svm import SVC
```

```python
dataset = pd.read_csv('./BankChurners.csv')
df = copy.deepcopy(dataset)
df
```

```python
col_1 = df['Attrition_Flag']
col_12 = df.iloc[:,2:29]
x= pd.DataFrame(col_12)  
y= pd.DataFrame(col_1)
print(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
kf = KFold(n_splits=10, shuffle=True, random_state=15)
test_errors = []
for train_index, test_index in kf.split(df):
    X_train, X_test = x.values[train_index], x.values[test_index]
    y_train, y_test = y.values[train_index], y.values[test_index]
    clf = SVC(kernel = 'rbf', random_state = 0)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    test_errors.append(precision_score(y_test, y_pred))
print(test_errors)
```

```python
test_errors_1=[]
for train_index, test_index in kf.split(df):
    X_train, X_test = x.values[train_index], x.values[test_index]
    y_train, y_test = y.values[train_index], y.values[test_index]
    clf1 = SVC(kernel = 'linear', random_state = 0)
    clf1.fit(X_train, y_train)
    y_pred = clf1.predict(X_test)
    test_errors_1.append(precision_score(y_test, y_pred))
print(test_errors_1)
```

```python
test_errors_2=[]
for train_index, test_index in kf.split(df):
    X_train, X_test = x.values[train_index], x.values[test_index]
    y_train, y_test = y.values[train_index], y.values[test_index]
    clf2 = SVC(kernel = 'poly', random_state = 0)
    clf2.fit(X_train, y_train)
    y_pred = clf2.predict(X_test)
    test_errors_2.append(precision_score(y_test, y_pred))
print(test_errors_2)
```

```python
print(np.mean(test_errors), np.mean(test_errors_1), np.mean(test_errors_2))
```
